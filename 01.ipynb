```python
title_list = list(set(title_list))
```

```python
keyword = 'tiktok'.lower()
count = 0
sorted_result = list()
for title in title_list:
    if keyword in title.lower():
        sorted_result.append([count, title.lower().count(keyword)])
    count += 1
sorted_result.sort(key=lambda x: x[1], reverse=True)
```
